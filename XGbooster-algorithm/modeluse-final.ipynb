```python
chunkrows=10000

import time
start_time = time.time()
import xgboost as xgb

import pandas as  pd

import numpy as np
import pandas as pd
import numpy as np
import joblib
np.set_printoptions(suppress=True)
out_path=""

#X=pd.read_csv('X.csv',low_memory=False)
# reader=pd.read_table('203.desc', low_memory=False,chunksize=10000)
chunk=pd.read_csv('203.data1',low_memory=False,nrows=chunkrows) 
hay=pd.read_csv('200k.hay3')#,low_memory=False,nrows=chunkrows)
chunk=chunk.append(hay)
```

```python
name=chunk['A']
# chunk=chunk.drop('A',axis=1)
model = joblib.load('model4.model')
droplist=list(set(chunk.columns) - set(model.feature_names))
i=0
for column in droplist:
    chunk=chunk.drop(column,axis=1)
```

```python
t=chunk.values    
t2=chunk.columns
t2
```

```python
chunk = xgb.DMatrix(data=t,feature_names=t2)
```

```python
output=[]
pred=model.predict(chunk,output_margin=True)
print(pred)
output=pd.DataFrame(output)
output['Name']=name
pred = [0 if item<0.7 else 1 for item in pred]
output['prediction']=pred
output=output.drop('Name',axis=1)
output
```

```python
output.to_csv('pred_from_model_use.csv')
```
